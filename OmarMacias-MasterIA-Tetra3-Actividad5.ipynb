**Pregunta 2**

Investiga la manera en que se realiza el análisis de discriminarte lineal Python y utiliza este conocimiento para levar a cabo la reducción de la dimensión y la representación gráfica del primer conjunto de datos construido en la pregunta anterior.

Para este caso se encontró que hay una librería de sklearn que permite realizar el análisis de discriminante lineal y se presenta a continuación.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

En el primer ejercicio se definieron los componentes en tres, se usará el LDA para reducirlo a dos componentes.
Separando los datos en train y test.

```python
yl = yl.astype(int) # esto se hizo porque se recibía un error al procesar el lda por la columnas de etiquetas que eran float y no int
x_train, x_test, y_train, y_test = train_test_split(Xl, yl, test_size=0.2, random_state=0) #Se usa para separar los valores de entrenamiento y prueba
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
print("Las dimensiones de los datos de prueba son: ", x_train.shape)
```

```python
lda = LDA(n_components=2)
x_train = lda.fit_transform(x_train, y_train)
x_test = lda.transform(x_test)
```

```python
print("Las dimensiones de el dataset generado son: ",x_train.shape)
print("Las dimensiones del dataset de prueba es: ",x_test.shape)
```

Generando la gráfica del dataset de entrenamiento.

```python
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, cmap='viridis')
plt.show()
```

**Pregunta 3**
Utilizando la librería Scikit-learn, investiga cómo se implementan las métricas del error cuadrático medio y del coeficiente de determinación  para evaluar la calidad de un modelo de aprendizaje supervisado.

Evaluando la calidad del modelo de aprendizaje tomando como ejemplo el el modelo de reducción reciente de LDA.

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
lda = lda.fit(x_train, y_train)
#Prediciendo los valores de los datos de prueba
y_pred = lda.predict(x_test)

#Calculando el error cuadrático medio
mse = mean_squared_error(y_test, y_pred)

#Calculando la raíz cuadrada del error cuadrático medio
rmse = np.sqrt(mse)
print("El error cuadrático medio es: ", str(mse))
print("La raíz cuadrada del error cuadrático medio es: ", str(rmse))
```

Calculando el coeficiente de determinación.

```python
r2 = lda.score(x_test, y_test)
print("El coeficiente de determinación es: ", str(r2))
```

Como podemos notar el coeficiente de determinación es bajo por lo que los valores de los datos de prueba no son muy aproximados.
