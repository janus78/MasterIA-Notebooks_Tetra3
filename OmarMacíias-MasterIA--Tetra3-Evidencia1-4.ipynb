Para este ejercicio se tomaron los datos de enfermos COVID del primer semestre de 2021 que presentaron cuadros de neumonía en la república mexicana.
Para motivos de este ejercicio se tomó una muestra más pequeña de ese set de datos ya que el archivo original ocupa 600 megas de espacio en disco.
La muestra que se tomó es de 20649 registros.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Cargando el dataset y obteniendo los valores de la columna "edad" y "sexo"

```python
df = pd.read_csv('CovidNeu1erSemestre2021.csv')
edad_lista = df['edad'].astype(int)
print(edad_lista)
sexo_lista = df['sexo'].astype(str) #0 es hombre, 1 es mujer
print(sexo_lista)
```

Creando el histograma de edades

```python
frecuencia, edad, ignorar = plt.hist(edad_lista, 30)
plt.title('Histograma de edades COVID con neumonía')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
plt.show()
```

Creando el histograma de sexos

```python
frecuencia, sexo, ignorar = plt.hist(sexo_lista, 2)
plt.title('Histograma de sexos COVID con neumonía')
plt.xlabel('Sexos')
plt.ylabel('Frecuencia')
plt.show()
```

Podemos notar claramente que el histograma que contiene las edades tiene una distribución normal en cambio el que tiene ambos sexos no presenta esta distribución, por lo que el cálculo de la media, la mediana y la desviación estándar  se realizará sobre las edades.

```python
media = np.mean(edad_lista)
print('Media: ', media)
mediana = np.median(edad_lista)
print('Mediana: ', mediana)
desviacion_estandar = np.std(edad_lista)
print('Desviación estándar: ', desviacion_estandar)
```

Creando la gráfica de la distribución gaussiana que se infiere de las edades.

```python
plt.title('Distribución gaussiana de edades COVID con neumonía')
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
frecuencia, temperatura, ignorar = plt.hist(edad_lista, 30, density=True)
plt.plot(edad_lista, 1/(desviacion_estandar * np.sqrt(2 * np.pi)) * np.exp( - (edad_lista - media)**2 / (2 * desviacion_estandar**2) ), 'ro')
plt.show()
```
