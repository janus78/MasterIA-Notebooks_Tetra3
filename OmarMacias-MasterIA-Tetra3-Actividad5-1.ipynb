Se encontró que usando la librería de sklearn se pueden generar diferentes dipos de datasets lineales o no lineales definiendo la cantidad de atributos y la cantidad de muestras.
Esta librería es la que se utiliza en la actividad.

**A. Conjunto de datos de tres características linealmente relacionados.**

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

Generando el dataset de regresión con tres características linealmente relacionadas.

```python
X, y = make_regression(n_samples=400, n_features=3, noise=0.1)
```

Creando la gráfica de los datos.

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
ax.scatter3D(X[:, 0], X[:, 1], X[:, 2], c=y, cmap='viridis')
ax.view_init(elev=30, azim=35)
plt.show()
```

**B. Generación del dataset con cuatro características relacionados de forma completa no lineal.**
Se menciona que usando el generador make_blobs de sklearn se puede generar un tipo de set de datos no lineal.
Al ser un dataset de 4 características no se graficará en este caso.

```python
#importando parte de la librería que genera este tipo de dataset
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_features=4, centers=4, cluster_std=10, random_state=11)
print("Las dimensiones de el dataset generado son: ",X.shape)
print("El contenido del dataset es:\n",X)
```
